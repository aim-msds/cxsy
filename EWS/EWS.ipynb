# Using Early Warning Signals on Real Dataset

* ONI (Oceanic Ni√±o Index)
* Rain 
* NIA (National Irrigation Administration)
* MWSS (Metropolitan Waterworks and Sewerage System)
* Dam 

```python
import sys
import pandas as pd

import matplotlib.pylab as plt

from ews import *
```

```python
full_data = pd.read_csv('Angat_dataset_raw.csv')
full_data.head()
```

```python
data = full_data.copy()
data_clean = data.dropna().reset_index(drop=True)
data_clean.head()
```

```python
for column in data_clean.columns:
    if column != 'Date':
        data_clean[[column, 'Date']].plot(x='Date', rot=20, legend=False, title=column)
```

## Conditional Heteroskedasticity
### Dam

```python
# data_clean
data_clean_ = data_clean.copy()
data_clean_['Date'] = pd.to_datetime(data_clean_['Date'])
data_clean_['Time'] = (data_clean_['Date'] - min(data_clean_['Date'])).dt.days
```

```python
dic = {'data': data_clean_.Dam, 
       'time' : data_clean_['Time'],
      'title': 'Dam',
      'method_name': 'CH'}

plot(data_clean, dic)
```

### Rain

```python
dic = {'data': data_clean_.Rain, 
       'time' : data_clean_['Time'],
      'title': 'Rain',
      'method_name': 'CH'}

plot(data_clean, dic)
```

### NIA

```python
dic = {'data': data_clean_.NIA, 
       'time' : data_clean_['Time'],
      'title': 'NIA',
      'method_name': 'CH'}

plot(data_clean, dic)
```

### ONI

```python
dic = {'data': data_clean_.ONI, 
       'time' : data_clean_['Time'],
      'title': 'ONI',
      'method_name': 'CH'}

plot(data_clean, dic)
```

### MWSS
